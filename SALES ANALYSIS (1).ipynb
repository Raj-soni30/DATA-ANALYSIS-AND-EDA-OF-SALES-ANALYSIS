```python
def seasons(Order_Month):
    if '02' <= Order_Month <= '04' :
        return 'Summer'
    elif '05' <= Order_Month <= '07':
        return 'Rainy'
    elif '08' <= Order_Month <= '10':
        return 'Autumn'
    else:
        return 'Winter'
```

```python
df['seasons'] = df['Order_Month'].apply(seasons)
```

```python
df.sample(5)
```

```python
df.drop('seasons',axis = 1 , inplace = True)
```

```python
seasons_wise_Sales = df.groupby('seasons')['Sales'].sum().reset_index()
seasons_wise_Sales
```

```python
plt.pie(x = seasons_wise_Sales['Sales'] , labels = seasons_wise_Sales['seasons'] , autopct = '%0.1f%%' , shadow = True , colors = ['#009596','#005F60','#A2D9D9','#8BC1F7'])
plt.title('SEASONS WISE SALES PERCENTAGE')
plt.show()
```
